<b>Background</b>

When you have a set of data, most of the time you find the mean of the data. The mean isn't always an accurate showing of the data. Many times you want to know how far off from the average the data actually is. To measure the amount of variation from the average you would find the <a href="http://en.wikipedia.org/wiki/Standard_deviation"> standard deviation</a>.

To find the standard deviation you need to know the mean value and each data point value.


\begin{equation*} stddev = \sqrt{(x_1 - meanvalue)^2)+(x_2 - meanvalue)^2)+ ... +(x_n - meanvalue)^2)} \end{equation*}

<b>Background</b>

$\chi^2$ or <a href="http://en.wikipedia.org/wiki/Chi-squared_distribution">chi square</a> is a statistical test used to compare observed data with the expected data of a specific hypothesis. Chi square can be used to find the best-fit line of data points.

\begin{equation*} \chi^2  = \sum_{i} \frac{(y(x_i) - y_i)^2}{ \theta_{y_i}^2} \end{equation*}
